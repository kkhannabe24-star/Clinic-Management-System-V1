```python
# --------------------------------------------------
# Generate Dates
# --------------------------------------------------

dates = pd.date_range(
    start="2025-01-01",
    end="2025-12-31",
    freq="D"
)

# --------------------------------------------------
# Generate Daily Patient Counts
# --------------------------------------------------

daily_count = []

day_type = ["Slow", "Average", "Busy"]

for _ in dates:

    d = np.random.choice(
        day_type,
        p=[0.2, 0.5, 0.3]
    )

    if d == "Slow":
        count = np.random.randint(5, 10)

    elif d == "Average":
        count = np.random.randint(10, 20)

    else:
        count = np.random.randint(20, 35)

    daily_count.append(count)

# --------------------------------------------------
# Generate Visits
# --------------------------------------------------

all_patients = patients_df["patient_id"].tolist()

visits = []

visit_counter = 1

for date, patient_count in zip(dates, daily_count):

    today_patients = np.random.choice(
        all_patients,
        size=patient_count,
        replace=False
    )

    for patient in today_patients:

        # --------------------------
        # Consultation Fee
        # --------------------------

        consultation_fee = 400

        # Complimentary visit (~3%)
        if np.random.random() < 0.03:
            consultation_fee = 0

        # --------------------------
        # Medicine Amount
        # --------------------------

        medicine_amount = np.random.choice(
            [0, 50, 100, 150, 200, 250, 300, 400],
            p=[0.05, 0.10, 0.20, 0.20, 0.20, 0.15, 0.08, 0.02]
        )

        # --------------------------
        # Service Amount
        # --------------------------

        service_amount = 0

        if np.random.random() < 0.15:

            service_amount = np.random.choice(
                [100, 150, 250, 300, 600],
                p=[0.30, 0.25, 0.20, 0.15, 0.10]
            )

        # --------------------------
        # Payment Mode
        # --------------------------

        payment_mode = np.random.choice(
            ["CASH", "ONLINE"],
            p=[0.65, 0.35]
        )

        # --------------------------
        # Create Visit Record
        # --------------------------

        visits.append([
            visit_counter,
            patient,
            date,
            consultation_fee,
            medicine_amount,
            service_amount,
            payment_mode
        ])

        visit_counter += 1

# --------------------------------------------------
# Create DataFrame
# --------------------------------------------------

visits_df = pd.DataFrame(
    visits,
    columns=[
        "visit_id",
        "p_id",
        "visit_date",
        "consultation_fee",
        "medicine_amount",
        "service_amount",
        "payment_mode"
    ]
)

# Preview
visits_df
```

```python
# ==================================================
# MEDICINE PURCHASED DATA GENERATION
# ==================================================

purchases = []

purchase_id = 1

vendors = [
    "ABC Pharma",
    "HealthCare Distributors",
    "MediPlus",
    "Sharma Medical Agency",
    "LifeCare Suppliers"
]

high_frequency = [
    2,   # Paracetamol
    3,   # Aceclofenac
    4,   # Diclofenac
    5,   # Etoricoxib
    14,  # Calcium Citrate
    15,  # Shelcal 500
    16,  # Vitamin D3
    21,  # Volini Gel
    22   # Dynapar Gel
]

for medicine in medicine_df["medicine_id"]:

    # ---------------------------------
    # High-use medicines purchased more
    # frequently than rare medicines
    # ---------------------------------

    if medicine in high_frequency:

        num_orders = np.random.randint(25, 40)

    else:

        num_orders = np.random.randint(10, 20)

    purchase_dates = np.random.choice(
        dates,
        size=num_orders,
        replace=False
    )

    for date in purchase_dates:

        # ---------------------------------
        # High-use medicines bought
        # in larger quantities
        # ---------------------------------

        if medicine in high_frequency:

            quantity_strips = np.random.randint(
                300,
                800
            )

        else:

            quantity_strips = np.random.randint(
                50,
                250
            )

        tablets_per_strip = medicine_df.loc[
            medicine_df["medicine_id"] == medicine,
            "tablets_per_piece"
        ].iloc[0]

        if pd.notna(tablets_per_strip):

            quantity_purchased = (
                quantity_strips *
                tablets_per_strip
            )

        else:

            quantity_purchased = quantity_strips

        vendor_name = np.random.choice(vendors)

        purchase_price = medicine_df.loc[
            medicine_df["medicine_id"] == medicine,
            "purchase_price_per_unit"
        ].iloc[0]

        total_purchase_cost = (
            quantity_purchased *
            purchase_price
        )

        purchases.append([
            purchase_id,
            medicine,
            date,
            quantity_purchased,
            total_purchase_cost,
            vendor_name
        ])

        purchase_id += 1

# ==================================================
# Create DataFrame
# ==================================================

medicine_purchased_df = pd.DataFrame(
    purchases,
    columns=[
        "purchase_id",
        "medicine_id",
        "purchase_date",
        "quantity_purchased",
        "total_purchase_cost",
        "vendor_name"
    ]
)

medicine_purchased_df.head()

```

```python
# ==================================================
# MEDICINE DISPENSED DATA GENERATION
# ==================================================

dispensed = []

dispense_id = 1

high_frequency = [
    2,   # Paracetamol
    3,   # Aceclofenac
    4,   # Diclofenac
    5,   # Etoricoxib
    14,  # Calcium Citrate
    15,  # Shelcal 500
    16,  # Vitamin D3
    21,  # Volini Gel
    22   # Dynapar Gel
]

all_medicines = medicine_df["medicine_id"].tolist()

for visit_id in visits_df["visit_id"]:

    # number of medicines prescribed
    medicine_count = np.random.choice(
        [0, 1, 2, 3, 4],
        p=[0.05, 0.20, 0.40, 0.25, 0.10]
    )

    if medicine_count == 0:
        continue

    selected_medicines = []

    for _ in range(medicine_count):

        # 70% chance of selecting common medicines
        if np.random.random() < 0.40:

            medicine = np.random.choice(
                high_frequency
            )

        else:

            medicine = np.random.choice(
                all_medicines
            )

        selected_medicines.append(medicine)

    # remove duplicates
    selected_medicines = list(
        set(selected_medicines)
    )

    for medicine in selected_medicines:

        unit_type = medicine_df.loc[
            medicine_df["medicine_id"] == medicine,
            "unit_type"
        ].iloc[0]

        if unit_type in ["TABLET", "CAPSULE"]:

            quantity_given = np.random.choice(
                [5, 10, 15, 20],
                p=[0.20, 0.40, 0.30, 0.10]
            )

        elif unit_type in ["TUBE", "SPRAY"]:

            quantity_given = 1

        else:

            quantity_given = np.random.choice(
                [1, 2]
            )

        dispensed.append([
            dispense_id,
            visit_id,
            medicine,
            quantity_given
        ])

        dispense_id += 1

medicine_dispensed_df = pd.DataFrame(
    dispensed,
    columns=[
        "dispense_id",
        "visit_id",
        "medicine_id",
        "quantity_given"
    ]
)

medicine_dispensed_df.head()
```

```python
purchased = (
    medicine_purchased_df
    .groupby("medicine_id")["quantity_purchased"]
    .sum()
)

dispensed = (
    medicine_dispensed_df
    .groupby("medicine_id")["quantity_given"]
    .sum()
)

inventory = purchased.subtract(
    dispensed,
    fill_value=0
)

inventory.sort_values().head(10)
```
